# Statistics via python

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Mean

```python
tempA=[70,50,80,120,50,75]

# calculate mean
meanA = np.mean(tempA)
print("mean:",meanA)
```

# Standard deviation

```python
# calculate standard deviation
stdA = np.std(tempA)
print("std:",stdA)
```

# coefficients of the standard deviation


```python
# calculate the coefficients of the standard deviation
coeff_stdA = stdA / meanA
print("coeff_std:",coeff_stdA)
```

# Variance

```python
# calculate variance
varA = np.var(tempA)
print("variance:",varA)
```

# Median 

```python
# calculate median
medianA = np.median(tempA)

print("median:",medianA)
```

# mode

```python
# calculate mode
from scipy import stats

modeA = stats.mode(tempA)

print("mode:",modeA)
```

# skewness

```python
# calculate skewness
dataA=[30, 40, 50, 60, 1000]
dataB=[1000, 800, 900, 700, 50]
skewA = stats.skew(dataA)
skewB = stats.skew(dataB)

print("skewness:",skewA)
print("skewness:",skewB)
plt.plot(dataA)
plt.plot(dataB)
plt.show()
```

# calculate the z-score


```python
# calculate the z-score
z_scoreA = stats.zscore(tempA)
result=z_scoreA
print("z_score:",result)
print(np.mean(result))
plt.plot(result)
```

# percentile Calculation

```python
Dataset=[10, 20, 30, 40, 50]
# percentile rank calculation
percentile_rank = np.percentile(Dataset, 30)
print("percentile_rank:",percentile_rank)
```

```python
Dataset1= [1, 2, 3, 10, 100]
# position of the 50th percentile
percentile_rank1 = np.percentile(Dataset1, 50)
print("percentile_rank of 50th:",percentile_rank1)

# position of the 75th percentile
percentile_rank2 = np.percentile(Dataset1, 75)
print("percentile_ranK of  75th:",percentile_rank2)

p=(percentile_rank2/6)*100
print("percentile_rank of 75th:",p)

```

# calculate the IQR of the data


```python
Dataset =[10, 12, 14, 15, 16, 18, 20, 100]
# calulate the IQR of the dataset
IQR = stats.iqr(Dataset)
print("IQR:",IQR)

# calculate the range of the dataset
rangeA = np.ptp(Dataset)

print("range:",rangeA)
```
